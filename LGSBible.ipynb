## **'Workers' Compensation'**

Workers' compensation stands as an exclusive remedy within the legal framework, designed to shield employers from lawsuits filed by employees for workplace injuries. It establishes a system where injured workers receive financial compensation and medical benefits without having to go through the complexities of traditional litigation.

## **'Occupational Safety and Health Administration (OSHA)'**

Established in 1972, the Occupational Safety and Health Administration (OSHA) is a federal agency with the primary objective of promoting workplace safety and health. In the context of employee rights, if an employer attempts to transfer an employee to a more hazardous part of the workplace, the employee has the legal right to refuse based on concerns for their safety. Importantly, the employer is prohibited from terminating the employee solely for refusing such a transfer.

## **'Fair Labor Standards Act (FLSA)'**

The Fair Labor Standards Act (FLSA) is a cornerstone of federal labor legislation, encompassing a comprehensive range of provisions related to wages and working conditions. Covering aspects such as child employment, wage standards, and the establishment of a minimum wage, the FLSA plays a pivotal role in ensuring fair labor practices. It's noteworthy that while there is a federal minimum wage, some jurisdictions may impose higher standards, and employees are entitled to the higher of the two rates based on their place of residence.

Overtime compensation, a critical component of the FLSA, dictates that employees must be paid at a rate of 1.5 times their regular pay for every hour worked beyond the standard 40 hours per week.

## **'Employment Laws for Minors'**

Legislation pertaining to the employment of minors sets specific restrictions on the types of work young individuals can undertake. For those under the age of 14, the permissible activities are limited to delivering newspapers, working on a family farm, or participating in the entertainment industry under adequate supervision.

## **'Family and Medical Leave Act (FMLA)'**

The Family and Medical Leave Act (FMLA) is a federal law that grants eligible employees up to 12 weeks of unpaid leave during a calendar year for specific family or medical reasons. Importantly, upon returning from leave, these employees are entitled to be reinstated to their original position or an equivalent one. It's crucial to note that this protection may not extend to key employees in high-ranking positions, such as those in the C-suite.

## **'Consolidated Omnibus Budget Reconciliation Act (COBRA)'**

COBRA, enacted in 1985, provides certain employees and their dependents the option to continue their healthcare coverage under their former employer's plan for a limited time after leaving employment. The standard coverage period is 18 months, but this can be extended to 26 months if the individual becomes disabled during the first 60 days of COBRA coverage.

## **'Workplace Privacy'**

To mitigate potential liability for invasion of privacy claims, employers are advised to implement clear policies regarding monitoring activities in the workplace. Providing notice to employees about the monitoring activities can serve as a preemptive measure, establishing transparency in the employer-employee relationship.

## **'Equal Employment Opportunity Commission (EEOC)'**

The Equal Employment Opportunity Commission (EEOC) is a federal agency tasked with overseeing and enforcing compliance with Title VII of the Civil Rights Act, which prohibits discrimination based on race, color, religion, sex, or national origin. Employees who believe they have been subjected to discrimination can file a complaint with the EEOC. If the EEOC finds merit in the complaint, it may take over the case; otherwise, it issues a "right to sue" letter, granting the employee permission to pursue legal action independently.

## **'Equal Pay Act'**

The Equal Pay Act, an amendment to the Fair Labor Standards Act, addresses disparities in wages based on gender. Notably, it has altered the statutes of limitations for equal pay claims, reducing the timeframe to 180 days from the most recent paycheck.

## **'Discrimination Laws'**

Federal laws prohibit discrimination in the workplace based on various factors, including gender, age, and national origin. The concept of reverse discrimination acknowledges instances where individuals from majority groups claim to be victims of discrimination based on their majority status.

## **'Sexual Harassment'**

Sexual harassment is explicitly prohibited under Title VII of the Civil Rights Act. Employers can invoke the Ellerth-Fargard defense to mitigate liability, provided they have robust sexual harassment policies and systems in place. For this defense to be successful, employers must demonstrate that the victim failed to follow proper channels to address and rectify the harassment.

## **'Pregnancy Discrimination'**

Pregnancy discrimination falls under the umbrella of gender discrimination. Employers are prohibited from discriminating against employees based on their pregnancy status, and affected individuals are entitled to legal remedies.

## **'Americans with Disabilities Act (ADA)'**

The Americans with Disabilities Act (ADA) extends protections to individuals with disabilities. Notably, individuals who have successfully completed rehabilitation for substance abuse are protected under the ADA.

## **'Age Discrimination in Employment Act (ADEA)'**

The Age Discrimination in Employment Act (ADEA) protects individuals aged 40 and older from discrimination based on age. To pursue a claim under the ADEA, an individual must be able to demonstrate that they are qualified for the job in question.

## **'Real Property'**

Real property refers to land and any improvements attached to or below the ground, including structures and subsurface rights. Ownership of real property grants individuals a comprehensive bundle of legal rights associated with the land.

## **'Concurrent Ownership'**

Concurrent ownership refers to situations where multiple individuals share ownership rights in a single piece of property. The two primary forms are tenancy in common, where each owner has an undivided interest, and tenancy by the entirety, a special form available only to spouses.

## **'Deeds'**

A deed is a legal instrument used to transfer ownership of real property from one party to another. There are different types of deeds, with general warranty deeds offering more extensive protection to the buyer than special warranty deeds. The protection period in a deed signifies the duration during which the seller is responsible for defects in title.

## **'Adverse Possession'**

Adverse possession is a legal concept that allows an individual to acquire title to a property by wrongfully and openly possessing it for a certain period, typically several years. The legal requirements for adverse possession vary by jurisdiction.

## **'Leases'**

Leases define the terms of a rental agreement and can take various forms, including fixed tenancy, tenancy at will, tenancy at sufferance, and periodic tenancy. Each type has distinct characteristics and legal implications.

## **'Easements'**

An easement grants a person the legal right to use another person's property for a specific purpose. Easements can be affirmative (granting a right to use the land) or negative (preventing the owner from using the land in a particular way).

## **'Consumer Protection Laws'**

Consumer protection laws operate at the federal, state, and local levels to safeguard consumers from unfair and deceptive trade practices. These laws promote product safety and ensure transparency in business transactions.

## **'Food and Drug Administration (FDA)'**

The Food and Drug Administration (FDA) is a federal agency responsible for enforcing the Food, Drug, and Cosmetic Act. Its primary role is to regulate the approval and sale of a wide range of products, including food, drugs, cosmetics, and medical devices.
